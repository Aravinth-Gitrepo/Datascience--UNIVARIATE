```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv ("Placement.csv")
```

```python
dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python
dataset["salary"].fillna(dataset['salary'].mode().iloc[0],inplace=True)
dataset
```

```python
#replace value ,, used inplace =True will replace value in dataset directly
dataset['salary'].fillna(0,inplace=True)
```

```python
dataset.isna().sum()
```

```python
#To remove the rows where the missing/na/null value present
dataset.dropna(inplace=True)
```

```python
dataset
```

```python
#replace with mean/median/mode
dataset["salary"].fillna(dataset['salary'].mean(),inplace=True)
```

```python
dataset
```

```python
dataset["salary"].fillna(dataset['salary'].median(),inplace=True)
```

```python
dataset
```

```python
dataset["salary"].fillna(dataset['salary'].mode(),inplace=True)
```

```python
dataset
```

```python
def quanQual(dataset):
            Quan=[]
            Qual=[]
            for columnName in dataset.columns:
                if dataset[columnName].dtypes=="O":
                    #print('Qual', columnName)
                    Qual.append(columnName)
                else:
                    #print('Quan',columnName)
                    Quan.append(columnName)
            return Quan, Qual
```

```python
Quan,Qual=quanQual(dataset)
Quan
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(dataset[Quan])
df=imp_mean.transform(dataset[Quan])
df = pd.DataFrame(df,columns = Quan)
#print(imp_mean.transform(dataset[Quan]))
```

```python
df
```

```python
two=[df,dataset[Qual]]
processeddata=pd.concat(two,axis=1)
```

```python
processeddata
```

```python
processeddata.to_csv("pre-processeddata.csv",index=False)
```
