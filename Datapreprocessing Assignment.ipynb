```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("kidney_disease.csv")
```

```python
dataset.isnull().sum()
```

```python
def quanQual(dataset):
            Quan=[]
            Qual=[]
            for columnName in dataset.columns:
                if dataset[columnName].dtypes=="O":
                    #print('Qual', columnName)
                    Qual.append(columnName)
                else:
                    #print('Quan',columnName)
                    Quan.append(columnName)
            return Quan, Qual
Quan,Qual=quanQual(dataset)
Quan
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(dataset[Quan])
df=imp_mean.transform(dataset[Quan])
df = pd.DataFrame(df,columns = Quan)
df
```

```python
for columnName in Quan:
    dataset[columnName].fillna(dataset[columnName].mode().iloc[0],inplace=True)
        
```
